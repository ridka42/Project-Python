```python
def replace_num(x):
    if x == '쿠킹실':
        return 1
    elif x == '포장실':
        return 2
    elif x == '계량실':
        return 3
    elif x == '충전실':
        return 4
    else:
        return 5
    
```

```python
product_booking_cooking_error['오류발생작업장명_num'] = product_booking_cooking_error['오류발생작업장명'].apply(lambda x: replace_num(x))
```
