## 랜덤 포레스트

```python
from sklearn.datasets import load_iris
import pandas as pd

iris_data = load_iris()
```

```python
x=pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
y=pd.DataFrame(iris_data.target, columns=['class'])
```

```python
df=pd.concat([x,y], axis=1)
df.to_csv('iris_data.csv')
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train_sc = scaler.transform(x_train)
x_test_sc = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(x_train_sc, y_train.values.ravel())
# .values.ravel() 안하면 y 때매 오류
# y를 시리즈형식이 아닌 데이터프레임 형식으로 넣었기 때문
```

```python
y_predict = clf.predict(x_test_sc)
y_predict
```

```python
clf.score(x_test_sc, y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, y_predict, average='micro')
# average의 디폴트가 binary라 바꿔줘야함 현재는 3개의 클래스
# micro는 3개의 평균
# None은 전부 나온다.
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
```

## 그래디언트 부스팅

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(random_state=0)
gb.fit(x_train_sc, y_train.values.ravel())
```
